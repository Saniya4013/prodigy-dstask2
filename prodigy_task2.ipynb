**1 IMPORT LIBRARIES**

```python
from warnings import filterwarnings
filterwarnings(action='ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(context='notebook',style='whitegrid',palette='muted')
```

**2.SHAPE OF DATA**

```python
df=pd.read_csv("/content/train.csv",index_col='PassengerId')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(columns='Cabin').isna().mean()
```

```python
df.drop(columns='Cabin').dropna(subset=['Embarked'])
```

```python
df['Age']=df['Age'].round()
df['Pclass']=df['Pclass'].map({1:'Upper',2:'Middle',3: 'Lower'})
df['Embarked']=df['Embarked'].map({"C":'Cherbourg','Q':'Queenstown','S':'Sounthampton'})
df['Survived']=df['Survived'].map({0:'Survived',1:' Not Survived'})
df['Sex']=df['Sex'].str.title()
categorical_columns=['Sex','Parch','SibSp','Pclass','Embarked','Survived']
df[categorical_columns]=df[categorical_columns].astype('category')
print(df)
```

```python
cols=['Name','Sex','Age','Parch','SibSp','Ticket','Pclass','Embarked','Fare','Survived']
df=df[cols]
```

**4.DATA EXPLORATION**

```python
sns.countplot(data=df,x='Survived')
plt.title('Survivals VS Non Survivals')
plt.show()
```

```python
def remove_outliers(df,col):
  Q1=df[col].quantile(0.25)
  Q3=df[col].quantile(0.75)
  IQR=Q3-Q1
  lower_bound=Q1-1.5*IQR
  upper_bound=Q3+1.5*IQR
  return df[(df[col]>=lower_bound)&(df[col]<=upper_bound)]
df=remove_outliers(df,'Age')
df=remove_outliers(df,'Fare')
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(15,7))
sns.boxplot(data=df,x="Age",hue='Survived',gap=0.1,ax=ax1)
ax1.set_xlabel('Age',fontsize=14)
ax2.set_title('Distribution of Fare')
plt.tight_layout()
plt.show();
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(15,7))
sns.boxplot(data=df,x="Age",hue='Survived',gap=0.1,ax=ax1)
ax1.set_xlabel('Age',fontsize=14)
ax2.set_title('Distribution of Fare')
plt.tight_layout()
plt.show();


fig,(ax1,ax2)=plt.subplots(1,2,figsize=(15,7))
sns.boxplot(data=df,x="Fare",hue='Survived',gap=0.1,ax=ax2)
ax1.set_xlabel('Fare',fontsize=14)
ax2.set_title('Distribution of Fare')
plt.tight_layout()
plt.show();
```

```python
cat_cols=['Sex','Parch','SibSp','Pclass','Embarked']
fig,axes=plt.subplots(2,3,figsize=(15,10))
for feature, ax in zip(cat_cols,axes.flatten()):
  sns.countplot(data=df,x=feature,hue='Survived',ax=ax)
  ax.set_title(f'Count of {feature}')
  ax.set_xlabel('')
  ax.set_ylabel('')
fig.delaxes(axes[1,2])
```
